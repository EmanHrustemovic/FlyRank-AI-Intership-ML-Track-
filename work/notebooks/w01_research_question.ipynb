## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

````python
import pandas as pd

# Clone repo first if not already done
import os
if not os.path.exists('/content/flyrank'):
    os.system('git clone https://github.com/EmanHrustemovic/FlyRank-AI-Intership-ML-Track- /content/flyrank')

df = pd.read_csv('/content/flyrank/data/raw/content_refresh_anonymized.csv')

total = len(df)
declining = (df['trend_direction'] == 'down').sum()
pct_declining = (df['trend_direction'] == 'down').mean() * 100
avg_impressions = df['impressions_90d'].mean()
avg_ctr = df['ctr'].mean()

print(f"Total pages: {total}")
print(f"Declining pages: {declining}")
print(f"% declining: {pct_declining:.1f}%")
print(f"Avg impressions (90d): {avg_impressions:.0f}")
print(f"Avg CTR: {avg_ctr:.3f}")
````
